```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
y = np.zeros(5)
x = np.linspace(0,1,5)
```

```python
fig = plt.plot(x,y,marker = 'o')
plt.show()
```

```python
def draw(x1,y1,x2,y2):
    plt.plot([x1,x2],[y1,y2],marker='o')
```

```python
import random
x = np.random.rand(5)
y= np.random.rand(5)

for i in range(0,5):
    draw(x[i%5],y[i%5],x[(i+1)%5],y[(i+1)%5])
```

```python
N= 10
x = np.linspace(0,1,N)
y= np.linspace(0,1,N)
```

```python
for i in range(0,N):
    for j in range(1,N):
        draw(x[j-1],y[i],x[j],y[i])
        draw(x[i],y[j-1],x[i],y[j])
```

```python
plt.figure(figsize=(N,N))
for i in range(0,N):
    for j in range(1,N):
        draw(x[j-1],y[i],x[j],y[i])
        draw(x[i],y[j-1],x[i],y[j])
        if i >=1:
            draw(x[i-1],y[j],x[i],y[j-1])
```
