```python
import numpy as np
N_test = [10,13,20,30,40,60]
h_test = [1/10,1/13,1/20,1/40,1/60]
log_h_test = np.log(h_test)
error = [0.0023591847750529223,
        0.0014057630956897473,
        0.0005974555841418471,
        0.00014985267309067409,
        6.664162868501065e-05
        ]
log_error = np.log(error)
import matplotlib.pyplot as plt
```

```python
plt.xlabel("h")
plt.ylabel("error L2")
plt.plot(h_test,error, marker = 'o')
```

```python
plt.figure(figsize = (6,6))
plt.xlabel("log h")
plt.ylabel("log (error L2)")
x = np.linspace(-4,-2.25,10)
y = 2*x
plt.plot(x,y,label = 'y=2x')
plt.legend()
plt.plot(log_h_test,log_error, marker = 'o')
```
