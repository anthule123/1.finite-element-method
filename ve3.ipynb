```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
M=32
def modif_function1(x, tam, r_tron,mul):
    r = np.linalg.norm(x-tam)
    if r >= r_tron:
        return 0
    else:
        return np.exp(mul*1/(r**2-r_tron**2))

x = np.outer(np.linspace(0,1, M), np.ones(M))
y = x.copy().T # transpose
z = np.zeros((M,M))

for i in range(0,M):
    for j in range(0,M):
            z[i][j] = modif_function1(np.array([x[i][j],y[i][j]]), np.array([1/2,1/2]), 1/2,1)
fig = plt.figure(figsize =(14, 9))
ax = plt.axes(projection ='3d')
 
# Creating plot
ax.plot_surface(x, y, z)
 
# show plot
plt.show()
```

```python
def d_modif_function1(x, tam, r_tron,mul):
    r = np.linalg.norm(x-tam)
    if r >= r_tron:
        return 0
    else:
        return -mul * modif_function1(x,tam,r_tron,mul)*(2*x[0]-2*tam[0])/(r**2- r_tron**2)**2
```

```python
for i in range(0,M):
    for j in range(0,M):
            z[i][j] = d_modif_function1(np.array([x[i][j],y[i][j]]), np.array([1/2,1/2]), 1/2,100)
fig = plt.figure(figsize =(14, 9))
ax = plt.axes(projection ='3d')
 
# Creating plot
ax.plot_surface(x, y, z)
 
# show plot
plt.show()
```
